```python
import numpy as np
import pandas as pd
import ipywidgets as widgets

hls_all_raw = pd.read_csv("/content/HSL_15022024231222956.csv")
print(hls_all_raw)
```

```python
print(hls_all_raw["Indicator"])
print("\n===========================================================\n")
hls_slice = pd.DataFrame(hls_all_raw, columns =["Country","Indicator","Type of indicator","Time","Value"])
print(hls_slice)
```

```python
hls_ls = hls_slice.loc[hls_all_raw["Indicator"] == "Life satisfaction"]
print(hls_ls)
print("\n===========================================================\n")
print("Total records:")
print(len(hls_ls))

print("\n===========================================================\n")
print("Total Unique Countries:")
print(len(hls_ls["Country"].unique()))

print("\n===========================================================\n")
print("Country List")
print(hls_ls["Country"].unique())
```

```python
hls_train = hls_ls.loc[hls_ls["Time"] == 2018]
hls_train = hls_train.loc[hls_ls["Type of indicator"] == "Average"]
print("\n===========================================================\n")
print("Total records:")
print(len(hls_train))

print("\n===========================================================\n")
print("Total Unique Countries:")
print(len(hls_train["Country"].unique()))

print("\n===========================================================\n")
print("Record:")
print(hls_train)
```

```python
weo_raw = pd.read_csv("/content/WEOOct2023all.xls", sep='\t', encoding='latin1')
print(weo_raw)
```

```python
# General government structural balance "GGSB_NPGDP"
weo_selected_measurement = weo_raw.loc[weo_raw['WEO Subject Code'].str.contains("GGSB_NPGDP", na=False)]
weo_selected_measurement_2021 = pd.DataFrame(weo_selected_measurement, columns=['Country', '2021'])

print(weo_selected_measurement_2021)
```

```python
merged_train_data = pd.merge(hls_train, weo_selected_measurement_2021, on="Country")
merged_train_data = merged_train_data.rename(columns={"Value": "Happiness Measurement", "2021": "General government structural balance"})
merged_train_data = pd.DataFrame(merged_train_data, columns=['Country','Happiness Measurement', 'General government structural balance'])

print(merged_train_data)

```

```python
import matplotlib.pyplot as plt
import sklearn.linear_model



X = np.c_[merged_train_data["General government structural balance"]]
Y = np.c_[merged_train_data["Happiness Measurement"]]
x = X.tolist()
y = Y.tolist()

# plot data
out1 = widgets.Output()
with out1:
  plt.scatter(x, y)
  plt.xlabel('General government structural balance')
  plt.ylabel('Happiness')
  plt.title("Data Plot")
  plt.show()

# fit linear model
model = sklearn.linear_model.LinearRegression()
model.fit(X, Y)

# plot predictions
predict_x = [x for x in range(901)]
predict_x = [[x/100] for x in predict_x]
predict_y = model.predict(predict_x)

out2 = widgets.Output()
with out2:
  plt.scatter(predict_x, predict_y)
  plt.scatter(x, y)
  plt.xlabel('General government structural balance')
  plt.ylabel('Happiness')
  plt.title("Prediction Line")
  plt.show()

display(widgets.HBox([out1,out2]))
```
